```python
import numpy as np
```

Array Creation

```python
arr = np.arange(1,21)
print(arr)
```

Sum of all elements in the array

```python
addition = np.sum(arr)
print(addition)
```

Mean of the elements 

```python
avg = np.mean(arr)
print(avg)
```

Median of the elements

```python
middle = np.median(arr)
print(middle)
```

Standard deviation

```python
std_dev = np.std(arr)
print(std_dev)
```

Index of Numbers greater then 10

```python
indices = np.where(arr > 10)
print(indices)
```

2D Array Creation

```python
arr2 = np.arange(1,17).reshape(4,4)
```

```python
print(arr2)
```

The transpose swaps rows and columns

```python
tr_arr2 = np.transpose(arr2)
print (tr_arr2)
```

```python
sum_row = np.sum(arr2 ,axis = 0)
print (sum_row)
```

Coloum wise Sum

```python
sum_colo = np.sum(arr2,axis = 1)
print (sum_colo)
```

Array With Random Numbers

```python
arr3_1 = np.random.randint(1, 21, size=(3, 3))
print(arr3_1)
```

```python
arr3_2 = np.random.randint(1, 21, size=(3, 3))
print(arr3_2)
```

Add element wise

```python
add_arr = arr3_1 + arr3_2
print(add_arr)
```

Substract element wise

```python
sub_arr = arr3_1 - arr3_2
print(sub_arr)
```

Multipication

```python
mul_arr = arr3_1 * arr3_2
print(mul_arr)
```

Dop Product

```python
pro_arr = np.dot(arr3_1,arr3_2)
print(pro_arr)
```
